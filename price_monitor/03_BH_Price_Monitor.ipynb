```python
# Dependencies
import requests
from bs4 import BeautifulSoup
from datetime import datetime
from pytz import timezone
import pandas as pd
import numpy as np
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
```

```python
# Make a reference to the books.csv file path
csv_path = "Resources/180826_B&H_ProductId.csv"

# Import the books.csv file as a DataFrame
bh_productid_df = pd.read_csv(csv_path, encoding="utf-8")
bh_productid_df
```

```python
for index, row in bh_productid_df.iterrows():
    url = row["producturl"]
    
    # Opening the page with chromedriver
    executable_path = {'executable_path': 'chromedriver.exe'}
    browser = Browser('chrome', **executable_path, headless=False)
    browser.visit(url)

    # Getting soup
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    try:
        price = soup.find('span', class_="itc-you-pay-price bold").text.replace('\t','').replace('$','').strip()
        price = float(price)
        
    except:
        price = None
    
    bh_productid_df.at[index, "B&H"] = price
```

```python
bh_result_df = bh_productid_df.drop(columns=["producturl"])
bh_result_df
```
